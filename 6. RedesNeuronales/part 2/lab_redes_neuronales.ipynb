## Desarrolle una Red Neuronal Artificial con Keras y Tensorflow para la compuerta XOR. 

Tenemos dos entradas binarias (1 ó 0) y la salida es 1 si una de las entradas es verdadera (1) y la otra falsa (0). 

Es decir que de cuatro combinaciones posibles, sólo dos tienen salida 1 y las otras dos serán 0, como vemos aquí:

* XOR(0,0) = 0
* XOR(0,1) = 1
* XOR(1,0) = 1
* XOR(1,1) = 0

### Cargue de las librerías

```python
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
```

### Se cargan las entradas

```python
datos_entrenamiento = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], "float32")

datos_entrenamiento
```

### Se cargan las salidas

```python
datos_salida = np.array([[0], [1], [1], [0]], "float32")
```

### Se programa la red neuronal

```python
# Crear el modelo de la red neuronal
model = Sequential()
model.add(Dense(4, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compilar el modelo
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Entrenar el modelo
model.fit(datos_entrenamiento, datos_salida, epochs=250)
```

```python
scores = model.evaluate(datos_entrenamiento, datos_salida)

print(model.metrics_names[1], scores[1]*100)
print(model.predict(datos_entrenamiento).round())
```

```python
# Predecir las salidas para los datos de entrenamiento y comparar con las salidas reales
predictions = model.predict(datos_entrenamiento).round()
print('Predictions:')
for i in range(len(datos_entrenamiento)):
    print(datos_entrenamiento[i], '-> Valor Predecido:', predictions[i], 'Valor Salida:', datos_salida[i])
```
